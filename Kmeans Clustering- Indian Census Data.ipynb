# Modelling

```python
# k-means
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
df_cluster.columns
```

```python
df_cluster_s = df_cluster[['Total Population', 'Illiterate', 'Graduate & above',
       'Percentage Illiterate', 'Percentage Graduate & above']]
df_cluster_s.head(2)
```

```python
# scalling all the columns so as to ensure they fall in the same range/
scaler = StandardScaler()

df_cluster_scaled = scaler.fit_transform(df_cluster_s)
df_cluster_scaled.shape
```

```python
kmeans = KMeans(n_clusters=3,max_iter=50,random_state=0)
kmeans.fit(df_cluster_scaled)
```

```python
kmeans.labels_
```

```python
df_cluster['cluster_id'] = kmeans.labels_
df_cluster.head()
```

```python
sns.scatterplot(y='Percentage Illiterate',x='cluster_id', data= df_cluster)
```

```python
sns.scatterplot(y='Percentage Graduate & above',x='cluster_id', data= df_cluster)

```

```python
sns.scatterplot(y='Percentage Graduate & above',x='Percentage Illiterate',hue='cluster_id', data= df_cluster)
```

```python
# elbow-curve/SSD
ssd = []
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(df_cluster_scaled)
    
    ssd.append(kmeans.inertia_)
    
# plot the SSDs for each n_clusters
# ssd
plt.plot(ssd)
```

```python
# BAased on silhoutte analysis we cn see that ideal clustering number is 3 or 4.
```

<b> Do we use elbow cure in real time projects when the features and data is hude? how reliable it is> </b> <br>
Not really, sometime elbow curve gives K=100 or 150 clusters which is really huge task to analyse and understand the nature of clusters as in decide to decide the usefulness from business stand point. <br>
Hence there is an another approach, where in we use hierarchal clustering first and save all the segmentation profiling. Upon which we analyse -as in which segment is making total sense from business stand point and how each k value is breaking down further. 

# Hierarchical Clustering

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import cut_tree
```

```python
scaler = StandardScaler()

df_cluster_scaled = pd.DataFrame(scaler.fit_transform(df_cluster_s))
df_cluster_scaled.shape
```

```python
df_cluster_scaled.head()
```

```python
df_cluster.head(3)
```

```python
# single linkage- wherein the clusters formed based on min distance

mergings = linkage(df_cluster_scaled, method="single",metric="euclidean")
dendrogram(mergings)
plt.show()
```

```python
mergings = linkage(df_cluster_scaled, method="complete",metric="euclidean")
dendrogram(mergings)
plt.show()
```

```python
# 3 clusters
cluster_labels = cut_tree(mergings, n_clusters=3).reshape(-1, )
cluster_labels
```

```python
df_cluster['cluster_labels'] =cluster_labels
df_cluster.head()
```
